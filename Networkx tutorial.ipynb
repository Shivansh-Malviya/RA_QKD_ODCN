# Adding Weights

```python
g = nx.Graph()
g.add_edge(1, 2, weight = 1100)
g.add_edge(2, 3, weight = 1600)
g.add_edge(1, 3, weight = 600)
g.add_edge(2, 8, weight = 1500)
g.add_edge(1, 4, weight = 1000)
g.add_edge(4, 5, weight = 600)
#g.add_edge(3, 4) #
```

```python
nx.draw_planar(g, with_labels = True)
```

```python
elarge = [ (u, v) for (u, v, d) in g.edges(data = True) if d["weight"] > 800]
esmall = [ (u, v) for (u, v, d) in g.edges(data = True) if d["weight"] <= 800]
```

```python
#pos = nx.spring_layout(g, seed = 9)
pos = nx.circular_layout(g)
```

```python
#nodes
nx.draw_networkx_nodes(g, pos, node_size =200)
```

```python
# edges
nx.draw_networkx_edges(g, pos, edgelist = elarge, width = 2)
nx.draw_networkx_edges(g, pos, edgelist = esmall, width = 2, alpha = 1, edge_color = 'b', style = "dashed") 
# Alpha is for opacity. 
```

## Labels

```python
# node labels
nx.draw_networkx_labels(g, pos, font_size = 20)

#edgelabels
edge_labels = nx.get_edge_attributes(g, "weight")
nx.draw_networkx_edge_labels(g, pos, edge_labels)
```

```python
def create_topology(edges):
  
    n = max(i[1] for i in edges)
    m = max(i[0] for i in edges)

    numNodes = max(n, m) # Calculates the number of nodes
     
    #print(numNodes)
    
    g = nx.Graph()

    #Nodes
    nodes = np.arange(1, numNodes)
    g.add_nodes_from(nodes)

    # Edges
    links = [(u, v) for (u, v, d) in edges]
    g.add_edges_from(links)

    #Weights
    weights = {(item[0], item[1]): item[2] for item in edges}

    #if(draw = True)
    #draw(g, pos, links)

    #return g

    #draw function
    #g = graph
    #nodes
    pos = nx.spring_layout(g, seed = 13) # seed =13
    
    nx.draw_networkx_nodes(g, pos, node_size =200)
    # edges
    nx.draw_networkx_edges(g, pos, edgelist = links, width = 2) 

    # node labels
    nx.draw_networkx_labels(g, pos, font_size = 8)

    #edge labels
    nx.draw_networkx_edge_labels(g, pos, weights)

    #if save_fig = True:
    #figname = input(print("Enter figure name : "))
    figname = "6-node1.png"
    plt.savefig(figname)

    #Configuraiton
    ax = plt.gca()
    ax.margins(0.08)
    plt.axis('off')
    plt.tight_layout()
    plt.show()

    return g
```

```python
 g = create_topology(edges)
```

```python
#clustering coefficient
clustering_coefficient = nx.average_clustering(g) #A high clustering coefficient implies tightly knit group
print(clustering_coefficient)
```

**Shortest Path** :

Sequence of nodes that minimizes the sum of edge weights

```python
# Shortest Path
shortest_path = nx.shortest_path(g, source = 1, target = 5)
print(shortest_path)
```

```python
# Betweenness centrality
betweenness_centrality = nx.betweenness_centrality(g)
print(betweenness_centrality)
```

```python
# Community Detection
communities = nx.algorithms.community.greedy_modularity_communities(g)

print(communities)
```
